```python
import pandas as pd 
import numpy as np 
```

```python
#read dataset using pandas
data = pd.read_csv("rain1.csv") 
```

```python
#to view top 5 rows of dataset
data.head()
```

```python
import sklearn as sk 
from sklearn.linear_model import LinearRegression
```

```python
# X is the input (contains all columns except the last one )
X = data.drop(['PrecipitationSumInches'], axis = 1) 
# Y is the output (contains last column of dataset)
Y = data['PrecipitationSumInches'] 
# reshape Y to 2d array from 1d array
Y = Y.values.reshape(-1, 1) 
```

```python
X.head()
```

```python
Y
```

```python
lr=LinearRegression()
```

```python
# fit input and output into linear regression model
lr.fit(X,Y)
```
